```python
import pandas as pd

demo = pd.read_csv('data/demografia.csv', encoding='latin-1', sep=';') # No identidica la ñ porque default usa utf-8, cambiamos el encoding y ya
```

```python
demo.head()
```

```python
demo.columns = ["edad", "sexo", "nacionalidad", "anno", "total"]
```

```python
demo.describe(include='all')
```

```python
demo.drop(['sexo', 'nacionalidad'], axis=1, inplace=True) # da error volver a ejecutarlo porque una vez se borra, no se puede volver a borrar
#demo.drop(['sexo', 'nacionalidad'], axis=1)
```

```python
print(demo)
```

```python
demo['total'] = demo['total']*1000 # Queremos quitarle el punto
```

```python
demo['total'] = demo['total'].astype(int)
```

```python
demo.describe() # Hemos quitado algunas columnas como sexo y nacionalidad, hemos quitado el punto en total y hemos cambiado el nombre de las comlumnas
```

```python
# calcular el numero de habitantes con 0 años desde 1998

calculo = sum(demo[(demo['edad'] == '0 años')]['total'])
#calculo = demo[demo['edad'] == '0 años']['total'].sum() # .sum es un metodo de numpy y sum() es un metodo de listas
# np.sum(demo['edad']== '0 años'['total']) # idem pero tenemos que importar numpy as np
# np.nansum(demo['edad']== '0 años'['total']) # idem pero tenemos que importar numpy as np
print(calculo)

calculo = sum(demo[~(demo['edad'] == '0 años')]['total']) # lo contrario

```

```python
# Cuantas personas con 14 años habia en el año 2000
calculo = demo[(demo['edad'] == '14 años')& (demo['anno'] == 2000)]['total'].sum()
print(calculo)
```

```python
# cuantas personas vivas habia en el ultimo año sin hardcodear el ultimo año
calculo = demo[demo['anno'] == demo['anno'].max()]['total'].sum()
print(calculo)
```

```python
# Datos solo del 2022
demo2022 = demo[demo['anno'] == 2022]
```

```python
# Porcentaje de cada año 
demo2022 = demo2022.copy() # Para que no aparezca el warning, nos avisa para que no sobreescriibamos el original, sin mas
total = demo2022['total'].sum()
demo2022['porcentaje'] = (demo2022['total']  / total) 
print(demo2022['porcentaje'].sum())
demo2022
```

```python
# Edad para el mayor porcentaje

maximaEdad = demo2022[demo2022['porcentaje'] == demo2022['porcentaje'].max()]['edad']
maximaEdadPorcentaje = demo2022[demo2022['porcentaje'] == demo2022['porcentaje'].max()]['porcentaje']
cantidad_de_la_misma = maximaEdadPorcentaje * total
print(maximaEdad, maximaEdadPorcentaje, cantidad_de_la_misma)

# SI queremos todo a la vez: 
maximaEdadTODO = demo2022[demo2022['porcentaje'] == demo2022['porcentaje'].max()] # Para obtener todo lo anterior pero en forma de tabla
maximaEdadTODO
```
