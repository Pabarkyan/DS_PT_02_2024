```python
# Cambiar el tipo (cast) de una columna
gen['Eólica'] = gen['Eólica'].astype("str")
gen['Eólica'] = gen['Eólica'].astype("int32")
```

```python
# Acceder a los nombres de las columnas
gen.columns
```

```python
# Cambiar los nombres de las colummnas
gen.columns = [x.lower() for x in gen.columns]
gen.columns
```

```python
# Crear una columna como una operación matemática sencilla de otra/s columns
gen['nuclearx10'] = gen['nuclear']*10
```

```python
# Crear una columna como una operación matemática sencilla de otra/s columns
gen['renovables'] = gen['eólica'] + gen['hidráulica'] + gen['solar fotovoltaica']
```

```python
# Eliminar columnas y sus parámetros inplace y errors
#gen = gen.drop(columns=["nuclearx10", "renovables"], errors="ignore")
gen.drop(columns=["nuclearx10", "renovables"], inplace=True, errors="ignore")
```

```python
# Operaciones matemáticas básicas 
print(gen['carbón'].sum())
print(gen['carbón'].mean())
print(gen['carbón'].median())
```

```python
# Convertir fechas
gen['hora'] = pd.to_datetime(gen['hora'])
```

```python
# Apply para operaciones complejas a cada fila

gen['dia'] = gen['hora'].apply(lambda x: x.day)
gen['import_export'] = gen['intercambios int'].apply(lambda x: "exportaciones" if x>0  else "importaciones")
print(gen[['hora','dia', 'intercambios int', 'import_export']].head(3)) 
print(gen[['hora','dia', 'intercambios int', 'import_export']].tail(3)) 
```

```python
# Mascaras/queries

# improtaciones de energía
print(gen[gen["import_export"] == "importaciones"]) 
print(gen.query("import_export == 'importaciones'")) 

```

```python
# nuclear > ciclo combinado y dia 17
print(gen[(gen["nuclear"] > gen["ciclo combinado"]) & (gen['dia'] == 17)]) 
```

```python
# nuclear > ciclo combinado y dia 17
# query no admite espacios, carácteres especiales... en los nombres de la columnas: renombrar
gen.rename(columns={'ciclo combinado': 'ciclo_combinado'}, inplace=True)
print(gen.query('(nuclear > ciclo_combinado)  & (dia == 17)')) 
```

```python
# Modificar parcialmente un dataframe
import numpy as np
gen.loc[gen['solar fotovoltaica'] == 0, 'solar fotovoltaica'] = np.nan
```

```python
# Group by
print(gen.groupby(['dia'])[['eólica']].sum())
print(gen.groupby(['dia'])[['eólica', 'nuclear']].mean())
print(gen.groupby(['dia']).aggregate({'eólica': 'sum', 'nuclear':'mean'}))
```
