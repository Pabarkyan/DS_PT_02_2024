```python
import pandas as pd
import numpy as np
import random
import datetime
```

```python
ids_original = pd.read_csv("bank_client_id.csv")
```

```python
ids = ids_original.copy()
ids.head()
```

```python
ids.info()
ids.describe(include='all') # Para sacarlas todas
```

```python
# Quitamos columna
ids.drop(columns=["Unnamed: 0"], inplace=True, errors="ignore") # Ignoramos errores, y no tenemos que volver a asignarlo a la variable con el inplace, es decir actualiza el mismo
```

```python
ids.head()
```

```python
ids.info() # Nos hemos ahorrado mas de 300 megas
```

```python
type(ids['open_date'][0]) # Vemos el tipo de la columna open date
```

```python
ids['open_date'] = pd.to_datetime(ids['open_date']) # Lo cambiamos a formato date time
ids['close_date'] = pd.to_datetime(ids['close_date'])
```

```python
ids.head() # NaT quiere decir Not a Time
```

```python
ids.describe()
```

```python
ids['year'] = ids['open_date'].apply(lambda x: x.year) # Para sacar el año de cada year
# Apply es para ejecutar linea por linea una transformacion de los datos, porque la columna como tal (series) no tiene el metodo year
# Pero cada uno de los elementos si.
```

```python
ids.head() # Error de 1970?
```

```python
ids.groupby(['year']).count() #cuanto elementos hay de cada elemento
# El groupby no cambia el ids pero crea una especie de objeto que agrupa coincidencias
```

```python
ids.groupby(['year'])[['close_date']].mean()
# Como observamos hay medias menores que el año en el q abrio
```

```python
ids[ids['close_date']<ids['open_date']] # Cuentas que se cerraron antes de abrirse, XD, asumimos que eestan mal, evidentemente
```

```python
ids.info()
```

```python
ids['year'] = ids['year'].astype('int16') # Para reducir el tamaño del archivo ponemos 16 pq sabemos que el año no supera el 32 mil
```

```python
ids.info()
```

```python
ids = ids[ids['year']!=1970] # Eliminamos el error de 1970
```

```python
ids = ids[ids['close_date'] > ids['open_date']]
```

```python
ids.info()
ids
```
