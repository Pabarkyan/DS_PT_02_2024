```python
# Con esta función leemos los datos y lo pasamos a un DataFrame de Pandas
def sql_query(query):

    # Ejecuta la query
    crsr.execute(query)

    # Almacena los datos de la query 
    ans = crsr.fetchall()

    # Obtenemos los nombres de las columnas de la tabla
    names = [description[0] for description in crsr.description]

    return pd.DataFrame(ans,columns=names)
```

Cuenta el número de employees por cada departamento (incluyendo el nombre en texto del departamento, no solo el código) de mayor número de employees a menor

```python
query = '''
SELECT department_name, COUNT(*) as n_empleados
FROM employees AS e
INNER JOIN departments as d
ON d.department_id = e.department_id
GROUP BY e.department_id
ORDER BY n_empleados DESC;
'''
sql_query(query)
```

Cuenta el número de employees por cada departamento y puesto de trabajo (incluyendo el nombre en texto del departamento y puesto, no solo el código)

```python
query = '''
SELECT department_name, job_title, COUNT(*) as n_empleados
FROM employees AS e
INNER JOIN departments AS d
ON d.department_id = e.department_id
INNER JOIN jobs AS j
ON e.job_id = j.job_id
GROUP BY e.job_id, e.department_id
ORDER BY n_empleados DESC
'''
sql_query(query)
```

Calcula el salario máximo, mínimo y medio por cada departamento

```python
query = '''
SELECT department_name, MAX(salary) AS salario_maximo, MIN(salary) AS salario_minimo, AVG(salary) AS salario_medio
FROM employees AS e
INNER JOIN departments AS d
ON d.department_id = e.department_id
GROUP BY e.department_id
'''
sql_query(query)
```

Cuenta el número de employees por cada departamento (incluyendo el nombre en texto del departamento, no solo el código) con más de 4 empleoyees (probar WHERE y HAVING)

```python
query = '''
SELECT department_name, COUNT(*) AS n_empleados
FROM employees AS e
INNER JOIN departments as d
ON d.department_id = e.department_id
GROUP BY e.department_id
HAVING n_empleados > 4
ORDER BY n_empleados DESC;
'''
sql_query(query)
```

Cuenta el número de employees por cada departamento y job (incluyendo el nombre en texto del departamento y job, no solo el código)

```python
query = '''
SELECT * FROM (
    SELECT department_name, COUNT(*) AS n_empleados -- COMENTARIO
    FROM employees AS e
    INNER JOIN departments as d
    ON d.department_id = e.department_id
    GROUP BY e.department_id
    ORDER BY n_empleados DESC) as t
WHERE n_empleados > 4;
'''
tabla = sql_query(query)
```
