```python
# Con esta función leemos los datos y lo pasamos a un DataFrame de Pandas
def sql_query(query):

    # Ejecuta la query
    crsr.execute(query)

    # Almacena los datos de la query 
    ans = crsr.fetchall()

    # Obtenemos los nombres de las columnas de la tabla
    names = [description[0] for description in crsr.description]

    return pd.DataFrame(ans,columns=names)
```

```python
query = '''
SELECT 
    employee_id, first_name, last_name, salary
FROM
    employees
WHERE
    department_id IN (SELECT 
            department_id
        FROM
            departments
        WHERE
            department_name = 'Marketing'
                OR department_name = 'Sales')
'''
sql_query(query)
```

Utiliza un inner join para extraer el nombre del departamento de cada empleado

```python
query = '''
    SELECT first_name, department_name
    FROM employees AS e
    INNER JOIN departments as d
    ON d.department_id = e.department_id
'''
sql_query(query)
```

Utiliza un inner join para extraer el nombre del departamento de cada empleado filtrando por departamentos 1, 2 y 3

```python
query = '''
    SELECT first_name, department_name
    FROM employees AS e
    INNER JOIN departments as d
    ON d.department_id = e.department_id
    WHERE e.department_id IN (1,2,3);
'''
sql_query(query)
```

Utiliza un left join para extraer la localización en cada país

```python
query = '''
    SELECT l.*, c.country_name
    FROM locations as l
    LEFT JOIN countries as c
    ON l.country_id = c.country_id
'''
sql_query(query)
```

Utiliza dos left join para extraer la localización, país y código de región

```python
query = '''
    SELECT location_id, country_id, country_name
    FROM locations as l
    LEFT JOIN countries as c
    ON l.country_id = c.country_id
'''
sql_query(query)
```

Cuenta el número de employees por cada departamento id
